### Exercício 6: Resolver uma Equação Quadrática

Escreva um script em Python que resolve a equação x^2 − 5x + 6 = 0 e imprime as raízes.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

def resolver_equacao_quadratica(a, b, c):
   """
   Resolve a equação quadrática ax² + bx + c = 0

   Parâmetros:
   a, b, c: Coeficientes da equação quadrática

   Retorna:
   Uma tupla com as raízes da equação
   """

   delta = b**2 - 4*a*c

   if delta < 0:
      return "Não há raízes reais"
   elif delta == 0:
      x = -b / (2*a)
      return (x, x)
   else:
      x1 = (-b + sqrt(delta)) / (2*a)
      x2 = (-b - sqrt(delta)) / (2*a)
      return (x1, x2)

a = 1
b = -5
c = 6

raizes = resolver_equacao_quadratica(a, b, c)

print(f"Equação: {a}x² - ({b})x + {c} = 0")
print(f"Raízes: x1 = {raizes[0]}, x2 = {raizes[1]}")

print("\nVerificação das raízes: ")
for i, raiz in enumerate(raizes, 1):
   resultado = a * raiz**2 + b * raiz + c
   print(f"Para x = {raiz}: {a}x({raiz})² + {b})x({raiz}) + {c} = {resultado:.10f}")

x = np.linspace(0, 5, 1000)
y = a * x**2 + b * x + c

plt.figure(figsize=(10, 6))
plt.plot(x, y, "b-", linewidth=2, label=f"f(x) = {a}x² + ({b})x + {c}")
plt.axhline(y=0, color="k", linestyle="-", alpha=0.3)
plt.axvline(x=0, color="k", linestyle="-", alpha=0.3)
plt.grid(True, linestyle="--", alpha=0.7)

plt.scatter(raizes, [0, 0], color="red", s=100, zorder=3, label="Raízes")
for raiz in raizes:
   plt.annotate(f"x = {raiz}", (raiz, 0), xytext=(0, 10), textcoords="offset points", ha="center", arrowprops=dict(arrowstyle="->", color="green"))

plt.title(f"Gráfico da Função Quadrática: {a}x² - ({b})x + {c}", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel(f"f(x)", fontsize=12)
plt.legend()
plt.ylim(-2, 10)
plt.tight_layout()
plt.show()

print("\nSolução alternativa usando numpy: ")
raizes_numpy = np.roots([a, b, c])
print(f"Raízes: x1 = {raizes_numpy[0]}, x2 = {raizes_numpy[1]}")

```

### Exercício 7: Equação Quadrática com Entrada do Usuário

Modifique o programa anterior para aceitar coeficientes a, b, e c de um usuário e resolver a equação quadrática correspondente.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

def resolver_equacao_quadratica(a, b, c):
   """
   Resolve a equação quadrática ax² + bx + c = 0

   Parâmetros:
   a, b, c: Coeficientes da equação quadrática

   Retorna:
   Uma tupla com as raízes da equação
   """

   if a == 0:
      if b == 0:
         if c == 0:
            return "Infinitas soluções (0 = 0)"
         else:
            return "Sem solução (equação impossível)"
      
      else:
         x = -c / b
         return f"Equação linear: x = {x}"

   delta = b**2 - 4*a*c
   if delta < 0:
      return "Não há raízes reais (discriminante negativo)"
   elif delta == 0:
      x = -b / (2*a)
      return (x, x)
   else:
      x1 = (-b + sqrt(delta)) / (2*a)
      x2 = (-b - sqrt(delta)) / (2*a)
      return (x1, x2)

def obter_coeficientes():
   """
   Obtém os coeficientes a, b, c do usuário com validação de entrada.
   """

   print("=" * 50)
   print("RESOLVEDOR DE EQUAÇÕES QUADRÁTICAS")
   print("Equação na forma: ax² + bx + c = 0")
   print("=" * 50)

   while True:
      try:
         a = float(input("Digite o coeficiente 'a' (coeficiente de x²): "))
         b = float(input("Digite o coeficiente 'b' (coeficiente de x): "))
         c = float(input("Digite o coeficiente 'c' (termo independente): "))
         return a, b, c
      except ValueError:
         print("Erro! Por favor, digite apenas números válidos.")
         print("Tente novamente.\n")

def formatar_equacao(a, b, c):
   """
   Formata a equação de forma legível
   """
   if a == 1:
      eq = "x²"
   elif a == -1:
      eq = "x²"
   else:
      eq = f"{a}x²"
   
   if b > 0:
      if b == 1:
         eq += " + x"
      else:
         eq += f" + {b}x"
   elif b < 0:
      if b == -1:
         eq += " - x"
      else:
         eq += f" - {-b}x"
   
   if c > 0:
      eq += f" + {c}"
   elif c < 0:
      eq += f"- {abs(c)}"
   
   return eq + " = 0"

def plotar_grafico(a, b, c, raizes):
   """
   Plota o gráfico da função quadrática
   """
   if a == 0:
      print("Não é possível pltar gráfico para equação não-quadrática.")
      return

   if isinstance(raizes, tuple) and len(raizes) == 2:
      x_min = min(raizes) - 2
      x_max = max(raizes) + 2
   else:
      x_min, x_max = -10, 20
   
   x = np.linspace(x_min, x_max, 1000)
   y = a * x ** 2 + b * x + c

   plt.figure(figsize=(10, 6))
   plt.plot(x, y, 'b-', linewidth=2, label=formatar_equacao(a, b, c).replace(' = 0', ''))
   plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
   plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
   plt.grid(True, linestyle="--", alpha=0.7)

   if isinstance(raizes, tuple) and len(raizes) == 2:
      plt.scatter(raizes, [0,0], color='red', s=100, zorder=3, label='Raízes')
      for i, raiz in enumerate(raizes):
         plt.annotate(f"x{i+1} = {raiz:.3f}", (raiz, 0), xytext=(0, -30 if i == 0 else 30), textcoords="offset points", ha="center", arrowprops=dict(arrowstyle="->", color="green"))

   plt.title(f"Gráfico da função f(x) = {formatar_equacao(a, b, c).replace(' = 0', '')}", fontsize=14)
   plt.xlabel('x', fontsize=12)
   plt.ylabel('f(x)', fontsize=12)
   plt.legend()
   plt.tight_layout()
   plt.show()

def main():
   """
   Função principal do programa
   """
   while True:
      a, b, c = obter_coeficientes()
      print(f"\nEquação: {formatar_equacao(a, b, c)}")
      resultado = resolver_equacao_quadratica(a, b, c)

      print("\n" + "=" * 30)
      print("RESULTADO: ")

      if isinstance(resultado, tuple):
         x1, x2 = resultado
         if x1 == x2:
            print(f"A equação possui uma raiz dupla: x = {x1}")
         else:
            print(f"A equação possui duas raízes reais: ")
            print(f"x1 = {x1}")
            print(f"x2 = {x2}")
         
         print("\nVerificação das raízes: ")
         for i, raiz in enumerate([x1, x2] if x1 != x2 else [x2], 1):
            resultado_verificao = a * raiz**2 + b * raiz + c
            print(f"Para x = {raiz}: f({raiz}) = {resultado_verificao:.10f}")
         plotar_grafico(a, b, c, resultado)
      
      else:
         print(resultado)
         if "linear" not in str(resultado):
            plotar_grafico(a, b, c, None)
      
      print("\n" + "=" * 50)
      continuar = input("Deseja resolver outra equação? (s/n): ").strip().lower()
      if continuar not in ['s', 'sim', 'y', 'yes']:
         print("Obrigado por usar o resolvedor de equação quadráticas!")
         break
      print()

if __name__ == "__main__":
   main()
```

### Exercício 8: Equação Quadrática com Discriminante

Para a equação quadrática ax^2 + bx + c = 0, calcule o discriminante e determine o número de raízes reais.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

def calcular_discriminante(a, b, c):
   """
   Calcula as raízes da equação quadrática ax² + bx + c = 0
   
   Parâmetros:
   a, b, c: Coeficientes da equação quadrática

   Retorna:
   O valor da discriminante (Delta = b² - 4ac) 
   """

   return b**2 - 4 * a * c

def analisar_discriminante(delta):
   """
   Analisa o discriminante e determina o número de raízes reais da equação quadrática.

   Parâmetros:
   delta: Valor do discriminante

   Retorna:
   Tupla contendo (numero_de_raizes, descrição)
   """
   if delta > 0:
      return (2, "Duas raízes reais distintas")
   elif delta == 0:
      return (1, "Uma raiz real (raiz dupla")
   else:
      return (0, "Nenhuma raiz real (duas raízes complexas)")

def calcular_raizes(a, b, c, delta):
   """
   Calcula as raízes da equação quadrática baseado no discriminante.
    
   Parâmetros:
   a, b, c: Coeficientes da equação
   delta: Discriminante
    
   Retorna:
   As raízes da equação ou None se não houver raízes reais
   """

   if delta > 0:
      x1 = (-b + sqrt(delta)) / (2 * a)
      x2 = (-b - sqrt(delta)) / (2 * a)
      return (x1, x2)
   elif delta == 0:
      x = -b / (2 * a)
      return (x, x)
   else:
      parte_real = -b / (2 * a)
      parte_imaginaria = sqrt(-delta) / (2 * a)
      return (f"{parte_real} + {parte_imaginaria}i", f"{parte_real} - {parte_imaginaria}i")

def formatar_equacao(a, b, c):
   """
   Formata a equação de forma legível.
   """

   eq_parts = []

   if a == 1:
      eq_parts.append("x²")
   elif a == -1:
      eq_parts.append("-x²")
   else:
      eq_parts.append(f"{a}x²")
   
   if b > 0:
      if b == 1:
         eq_parts.append(" + x")
      elif b == -1:
         eq_parts.append(f" + {b}x")
   elif b < 0:
      if b == -1:
         eq_parts.append(" - x")
      else:
         eq_parts.append(f" - {abs(b)}x")

   if c > 0:
      eq_parts.append(f" + {c}")
   elif c < 0:
      eq_parts.append(f" - {abs(c)}x")

   return "".join(eq_parts) + " = 0"

def plotar_grafico_com_discriminante(a, b, c, delta, raizes):
   """
   Plota o gráfico da função quadrática destacando a análise do discriminante.
   """
   vertice_x = -b / (2 * a)
   intervalo = max(10, abs(vertice_x) + 5)
   x = np.linspace(vertice_x - intervalo, vertice_x + intervalo, 1000)
   y = a * x**2 + b * x + c

   plt.figure(figsize=(12, 8))

   plt.subplot(2, 1, 1)
   plt.plot(x, y, 'b-', linewidth=2, label=formatar_equacao(a, b, c).replace(' = 0', ''))
   plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
   plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
   plt.grid(True, linestyle="--", alpha=0.7)

   vertice_y = a * vertice_x**2 + b * vertice_x + c
   plt.scatter(vertice_x, vertice_y, color='orange', s=100, zorder=3, label='Vértice')

   if delta > 0:
      plt.scatter(raizes[0], 0, color="red", s=100, zorder=3, label="Raiz 1")
      for i, raiz in enumerate(raizes, 1):
         plt.annotate(
            f"x{i} = {raiz:.3f}",
            (raiz, 0),
            xytext=(0, -30 if i == 1 else 30),
            textcoords="offset points",
            ha="center",
            arrowprops=dict(arrowstyle="->", color="green"),
         )
      plt.title(f"Δ = {delta} > 0: Parábola cruza o eixo x em dois pontos", fontsize=12, color="green")
   
   elif delta == 0:
      plt.scatter(raizes[0], 0, color="red", s=150, zorder=3, label="Uma raiz real(dupla)")
      plt.annotate(f"x = {raizes[0]:.3f}", (raizes[0], 0), xytext=(0, -30), textcoords="offset points", ha="center", arrowprops=dict(arrowstyle="->", color="gold"))

      plt.title(f"Δ = {delta} = 0: Parábola toca o eixo x em um ponto", fontsize=12, color="orange")
   
   else:
      plt.title(f"Δ = {delta} < 0: Parábola não cruza o eixo x", fontsize=12, color="red")

   plt.xlabel("x")
   plt.ylabel("f(x)")
   plt.legend()

   plt.subplot(2, 1, 2)
   plt.axis("off")

   info_text = f"""
   Análise do Discriminante (Δ):

   Equação: {formatar_equacao(a, b, c)}
   Coeficientes: a = {a}, b = {b}, c = {c}
   Discriminante (Δ) = b² - 4ac = ({b}² - 4*{a}*{c}) = {delta}

   Interpretação:
   """

   if delta > 0:
      info_text += f"• Δ = {delta} > 0 → DUAS RAÍZES REAIS DISTINTAS\n"
      info_text += f"• x₁ = {raizes[0]:.6f}\n"
      info_text += f"• x₂ = {raizes[1]:.6f}\n"
      info_text += "• A parábola cruza o eixo x em dois pontos diferentes.\n"
   elif delta == 0:
      info_text += f"• Δ = {delta} = 0 → UMA RAIZ REAL (RAIZ DUPLA)\n"
      info_text += f"• x = {raizes[0]:.6f}\n"
      info_text += "• A parábola toca o eixo x em um ponto.\n"
   else:
      info_text += f"• Δ = {delta} < 0 → NENHUMA RAIZ REAL\n"
      info_text += f"• Raízes complexas: {raizes[0]} e {raizes[1]}.\n"
      info_text += "• A parábola não cruza nem toca o eixo x.\n"

   plt.text(0.05, 0.95, info_text, transform=plt.gca().transAxes, fontsize=10, verticalalignment="top", fontfamily="monospace", bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgray", alpha=0.8))

   plt.tight_layout()
   plt.show()

def obter_coeficientes():
   """
   Obtém os coeficientes a, b, c do usuário.
   """
   print("=" * 60)
   print("ANÁLISE DO DISCRIMINANTE DE EQUAÇÕES QUADRÁTICAS")
   print("Equação na forma: ax² + bx + c = 0")
   print("=" * 60)

   while True:
      try:
         a = float(input("Digite o coeficiente 'a' (deve ser diferente de zero): "))
         if a == 0:
            print("Erro! O coeficiente 'a' deve ser diferente de zero para uma equação quadrática.")
            continue
         b = float(input("Digite o coeficiente 'b': "))
         c = float(input("Digite o coeficiente 'c': "))
         return a, b, c
      except ValueError:
         print("Erro!, Por favor, digite apenas números válidos.")

def main():
   """
   Função principal do programa
   """
   while True:
      a, b, c = obter_coeficientes()
      delta = calcular_discriminante(a, b, c)
      num_raizes, descricao = analisar_discriminante(delta)
      raizes = calcular_raizes(a, b, c, delta)

      print(f"\n" + "=" * 60)
      print("RESULTADOS DA ANÁLISE")
      print(f"Equação: {formatar_equacao(a, b, c)}")
      print(f"Discriminante (Δ) = {delta}")
      print(f"Número de raízes reais: {num_raizes}")
      print(f"Descrição: {descricao}")

      if num_raizes > 0:
         print(f"Raízes: {raizes[0]}" + (f" e {raizes[1]}" if raizes[0] != raizes[1] else " (dupla)"))
      else:
         print(f"Raízes complexas: {raizes[0]} e {raizes[1]}")
      
      plotar_grafico_com_discriminante(a, b, c, delta, raizes)

      print("\n" + "=" * 60)
      continuar = input("Deseja atuaizar outra equação? (s/n): ").strip().lower()
      if continuar not in ['s', 'sim', 'y', 'yes']:
         print("Obrigado por usar o analisador de discriminantes!")
         break
      print()

def exemplos_demonstracao():
   """
   Executa exemplos demonstrativos dos três casos do discriminante.
   """

   print("EXEMPLOS DEMONSTRATIVOS:")
   print("=" * 60)

   exemplos = [
      (1, -5, 6, "Δ > 0 (duas raízes reais)"),
      (1, -4, 4, "Δ = 0 (uma raiz real)"),
      (1, 0, 1, "Δ < 0 (nenhuma raiz real)")
   ]

   for a, b, c, descricao in exemplos:
      print(f"\n{descricao}")
      print(f"Equação: {formatar_equacao(a, b, c)}")

      delta = calcular_discriminante(a, b, c)
      num_raizes, desc = analisar_discriminante(delta)
      raizes = calcular_raizes(a, b, c, delta)

      print(f"Discriminante: {delta}")
      print(f"Resultado: {desc}")
      if num_raizes > 0:
         print(f"Raízes: {raizes}")
      else:
         print(f"Raízes complexas: {raizes}")

if __name__ == "__main__":
   exemplos_demonstracao()

   print("\n" + "=" * 60)
   input("Pressione Enter para começar a análise interativa...")

   main()
```

### Exercício 9: Cálculo de Logaritmos

Escreva um script que calcula o logaritmo de base 10 de um número fornecido pelo usuário. Certifique-se de tratar o caso em que o logaritmo não pode ser definido (números negativos e zero).

```python
import math

def calcular_logaritmo():
   """
   Calcula o logaritmo de base 10 de um número fornecido pelo usuário.
   Trata casos especiais onde o logaritmo não está definido.
   """
   print("=== Calculadora de Logaritmo Base 100 ===")
   print()

   while True:
      try:
         numero = float(input("Digite um número para calcular log10: "))
         if numero <= 0:
            if numero == 0:
               print("Erro: O logaritmo de zero não está difinido!")
               print("  Motivo: log10(0) tende a -∞")
            else:
               print(f"Erro: O logaritmo de {numero} não está difinido!")
               print("  Motivo: Logaritmo de npuemros negativos não existe nos números reais")
            print()

            continuar = input("Deseja tentar com outro número? (s/n): ").lower().strip()
            if continuar != 's':
               break
            print()
            continue
         resultado = math.log10(numero)
         print(f"log10({numero}) = {resultado:.6f}")
         print(f"   Verificação: 10^{resultado:.6f} = {10**resultado:.6f}")

         if numero == 1:
            print("Nota: log10(1) = 0 (qualquer base elevada a 0 = 1)")
         elif numero == 10:
            print("Nota: log10(10) = 1 (logartimo da própria base)")
         elif numero == 100:
            print("Nota: log10(100) = 2 (10² = 100)")
         elif numero == 1000:
            print("Nota: log10(1000) = 3 (10³ = 1000)")
         elif 0 < numero < 1:
            print("Nota: Números entre 0 e 1 têm logaritmo negativo")
         print()

         continuar = input("Deseja calcular outro logaritmo? (s/n): ").lower().strip()
         if continuar != "s":
            break
         print()
      except ValueError:
         print("Erro: Por favor, digite um número válido!")
         print()
         continuar = input("Deseja tentar novamente? (s/n): ").lower().strip()
         if continuar != "s":
            break
         print()
      except Exception as e:
         print(f"Erro inesperado: {e}")
         break
   print("Obrigado por usar a calculadora de logaritmos!")

def demonstrar_propriedades():
   """
   Demonstra algumas propriedades importantes dos logaritmos.
   """
   print("\n=== Propriedades dos Logaritmos Base 10 ===")
   print()

   a, b = 100, 1000
   
   print(f"Exemplos com a = {a} e b = {b}: ")
   print(f"log10({a}) = {math.log10(a)}")
   print(f"log10({b}) = {math.log10(b)}")
   print()

   print("Propriedades: ")
   print("log10(a x b) = log10(a) + log10(b)")
   print(f"  log10({a} x {b}) = {math.log10(a*b):.3f}")
   print(f"  log10({a}) + log10({b}) = {math.log10(a) + math.log10(b):.3f}")
   print()

   print(f"log10(a / b) = log10(a) - log10(b)")
   print(f"  log10({a} / {b}) = {math.log10(a/b):.3f}")
   print(f"  log10({a}) - log10({b}) = {math.log10(a)} - {math.log10(b):.3f}")
   print()

   print(f"log10(a^n) = n x log10(a)")
   n = 3
   print(f"   log10({a}³) = {math.log10(a**n):.3f}")
   print(f"   3 x log10({a}) = {n * math.log10(a):.3f}")

if __name__ == "__main__":
   calcular_logaritmo()
   print()
   mostrar_pop = input("Deseja ver exemplos das propriedades dos logaritmos? (s/n): ").lower().strip()
   if mostrar_pop == 's':
      demonstrar_propriedades()
```

### Exercício 10: Aplicação de Logaritmos

Utilize logaritmo para calcular o tempo necessário para que um investimento atinja um determinado valor, considerando uma taxa de crescimento contínua.

```python
import math

def calcular_tempo_investimento():
   """
   Calcula o tempo necessário para um investimento atingir um valor alvo
   usando crescimento contínuo e logaritmos naturais.

   Fórmula: A = P * e^(r*t)
   Resolvendo para t: t = ln(A/P) / r
   """
   print("=== Calculadora de Tempo de Investimento ===")
   print("Crescimento Contínuo: A = P x e (rxt)")
   print()

   while True:
      try:
         print("Digite os dados do investimento:")
         principal = float(input("Valor inicial (P): R$ "))
         if principal <= 0:
            print("O valor incial deve ser positivo!")
            continue

         valor_alvo = float(input("Valor alvo (A): R$ "))
         if valor_alvo < principal:
            print("O valor alvo deve ser maior que o valor incial!")
            continue
         
         taxa_anual = float(input("Taxa de juros anual (%): "))
         if taxa_anual <= 0:
            print("A taxa de juros deve ser positiva!")
            continue

         r = taxa_anual / 100

         print("\n" + "="* 50)
         print("RESULTADO DA ANÁLISE")
         print("=" * 50)

         razao = valor_alvo / principal
         tempo_anos = math.log(razao) / r

         tempo_meses = tempo_anos * 12
         tempo_dias = tempo_anos * 365.25

         print(f"\n DADOS DO INVESTIMENTO:")
         print(f"  Valor Inicial: R$ {principal:,.2f}")
         print(f"  Valor Alvo: R$ {valor_alvo:,.2f}")
         print(f"  Taxa Anual: {taxa_anual}% ({r:.4f} decimal)")

         print("\n TEMPO NECESSÁRIO:")
         print(f"  {tempo_anos:.2f} anos")
         print(f"  {tempo_meses:.2f} meses")
         print(f"  {tempo_dias:.2f} dias")

         print(f"\n CÁLCULO MATEMÁTICO:")
         print(f" - Fórmula: t = ln(A/P) / r")
         print(f" - t = ln({valor_alvo:,.0f} / {principal:,.0f}) / {r:.4f}")
         print(f" - t = ln({razao:.4f}) / {r:.4f}")
         print(f" - t = {math.log(razao):.4f} / {r:.4f}")
         print(f" - t = {tempo_anos:.4f} anos")

         valor_verificacao = principal * math.exp(r * tempo_anos)
         print(f"\n VERIFICAÇÃO:")
         print(f"  P x e^(rxt) = {principal:,.2f} x e^({r:.4f}x{tempo_anos:.4f})")
         print(f"  = {principal:,.2f} x {math.exp(r*tempo_anos):.4f}")
         print(f"  = R$ {valor_verificacao:,.2f}")
         print(f"  Diferença: R$ {abs(valor_verificacao - valor_alvo):.2f}")

         analisar_investimento(principal, valor_alvo, taxa_anual, tempo_anos)

         mostrar_tabela = input("\nDeseja ver a tabela de crescimento ano a ano? (s/n): ").lower().strip()
         if mostrar_tabela == "s":
            gerar_tabela_crescimento(principal, r, tempo_anos)
         
         print("\n" + "="*50)
         continuar = input("Deseja fazer outro cálculo? (s/n): ").lower().strip()
         if continuar != 's':
            break
         print()
      except ValueError:
         print("Erro: Digite apenas números válidos!")
      except Exception as e:
         print(f"Erro inesperado: {e}")
   
   print("\nObrigado por usar a calculadora!")

def analisar_investimento(principal, valor_alvo, taxa_anual, tempo_anos):
   """Fornece análises adicionais sobre o investimento."""
   print(f"\n ANÁLISE DO INVESTIMENTO:")

   rendimento_total = valor_alvo - principal
   rentabilidade_percentual = (rendimento_total / principal) * 100

   print(f" - Rendimento total: R$ {rendimento_total:,.2f}")
   print(f" - Rentabilidade percentual: {rentabilidade_percentual:.1f}%")

   valor_juros_compostos = principal * (1 + taxa_anual/100) ** tempo_anos
   diferenca = valor_alvo - valor_juros_compostos

   print(f"\n COMPARAÇÃO: ")
   print(f" - Crescimento Contínuo: R$ {valor_alvo:,.2f}")
   print(f" - Juros Compostos Anuais: R$ {valor_juros_compostos:,.2f}")
   print(f" - Diferença: R$ {diferenca:,.2f}")

   if tempo_anos < 1:
      classificacao = "Muito Rápido (menos de 1 ano)"
   elif tempo_anos < 5:
      classificacao = "Custo Prazo (1-5 anos)"
   elif tempo_anos < 15:
      classificacao = "Médio Prazo (5-15 anos)"
   else:
      classificacao = "Longo Prazo (mais de 15 anos)"
   
   print(f"  - Classificação: {classificacao}")


def gerar_tabela_crescimento(principal, taxa_decimal, tempo_total):
   """Gera uma tabela mostrando o crescimento do investimento."""
   print(f"\n TABELA DE CRESCIMENTO:")
   print("-" * 60)
   print(f"{"Ano":<8} {"Valor (R$)":<15} {"Crescimento":<15} {"% do Alvo"}")
   print("-" * 60)

   valor_alvo = principal * math.exp(taxa_decimal * tempo_total)
   anos_para_mostrar = min(int(tempo_total) + 1, 20)
   for ano in range(anos_para_mostrar + 1):
      if ano > tempo_total:
         break

      valor_atual = principal * math.exp(taxa_decimal * ano)
      if ano == 0:
         crescimento = "Inicial"
      else:
         crescimento_anual = valor_atual - (principal * math.exp(taxa_decimal * (ano-1)))
         crescimento = f"R$ {crescimento_anual:,.0f}"
      
      percentual_alvo = (valor_atual / valor_alvo) * 100

      print(f"{ano:<8} {valor_atual:>12,.2f}   {crescimento:<13} {percentual_alvo:>6.1f}%")

   if tempo_total % 1 != 0:
      valor_final = principal * math.exp(taxa_decimal * tempo_total)
      print(f"{tempo_total:<8.2f} {valor_final:>12,.2f}  {"Meta!":<13} {"100.0%":>7}")

def calcular_cenarios():
   """Calcula diferentes cenários de investimento."""
   print("\n=== ANÁLISE DE CENÁRIOS ===")

   try:
      principal = float(input("Valor inicial: R$ "))
      valor_alvo = float(input("Valor alvo: R$ "))

      if principal <= 0 or valor_alvo <= principal:
         print("Valores inválidos!")
         return
      
      print("\n TEMPO NECESSÁRIO COM DIFERENTES TAXAS:")
      print("-" * 50)
      print(f"{'Taxa Anual':<12}  {'Tempo (anos)':<15} {"Tempo (meses)"}")
      print("-" * 50)

      taxas = [3, 5, 7, 10, 12, 15, 18, 20]
      razao = valor_alvo / principal

      for taxa in taxas:
         r = taxa / 100
         tempo = math.log(razao) / r
         tempo_meses = tempo * 12

         print(f"{taxa}%{"":<9} {tempo:<15.2f} {tempo_meses:.1f} meses")
      
      print("-" * 50)
   
   except ValueError:
      print("Erro: Digite valores numéricos válidos!")

def exemplo_pratico():
   """Mostra um exemplo prático de aplicação."""
   print("\n=== EXEMPLO PRÁTICO ===")
   print("  Meta: Transformar R$ 10.000 em R$ 50.000")
   print("  Taxa: 8% ao ano (crescimento contínuo)")
   print()

   P = 10000
   A = 50000
   r = 0.08

   tempo = math.log(A/P) / r

   print(f"Cálculo:")
   print(f"t = ln({A} / {P}) / {r}")
   print(f"t = ln({A/P:.1f}) / {r}")
   print(f"t = {math.log(A/P):.4f} / {r}")
   print(f"t = {tempo:.2f} anos")
   print(f"t = {tempo * 12:.1f} meses")

   print(f"\n Para multiplicar seu dinheiro por 5x com taxa de 8% ao ano,")
   print(f"  você precisará de aproximadamente {tempo:.1f} anos!")

if __name__ == "__main__":
   while True:
      print("\n" + "="*60)
      print("  CALCULADORA DE INVESTIMENTOS COM LOGARITMOS")
      print("="*60)
      print("1. Calcular tempo de investimento")
      print("2. Análise de cenários (diferentes taxas)")
      print("3. Ver exemplo prático")
      print("4. Sair")
      print()

      opcao = input("Escolha uma opção (1-4): ").strip()

      if opcao == '1':
         calcular_tempo_investimento()
      elif opcao == '2':
         calcular_cenarios()
      elif opcao == '3':
         exemplo_pratico()
      elif opcao == '4':
         print("Obrigado por usar a calculadora!")
         break
      else:
         print("Opção inválida! Escolha entre 1-4.")
      
      input("\nPressione Enter para continuar...")
```
