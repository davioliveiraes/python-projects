```python
import pandas as pd

tabela_vendas = pd.read_excel(
    r"C:\Users\Davil\OneDrive\Documentos\Practice\Python\ExerciciosEmPython\enviar-email-automaticamente\Vendas.xlsx"
)
display(tabela_vendas)

total_gastos = tabela_vendas["Valor Final"].sum()
total_qtde = tabela_vendas["Quantidade"].sum()

preco_medio = total_gastos / total_qtde
print(f"O valor total de gastos foi: R${total_gastos:,.2f}")
print(f"O total da quantidade vendida foi: {total_qtde}")
print(f"O valor médio foi: R${preco_medio:,.2f}")
```
